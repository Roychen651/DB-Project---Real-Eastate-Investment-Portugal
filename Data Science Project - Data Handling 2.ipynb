```python
# show all the types of the details and the size of the table ( rows X cols ) , *before data handling

df_handling = pd.read_csv('Real_estate_data_Portugal.csv', header=0, sep=',') 
print('Real Estate columns:', df_handling.columns)
print('Real Estate dtypes:', df_handling.dtypes)
print('Real Estate shape:', df_handling.shape)
df_handling.head()
```

```python
# steps to convert str to int or float, first remove unnecessary characters
# split from the Address the division and the city to new columns.

df_handling['Price'] = df_handling['Price'].str.replace(' ', '')
df_handling['Price'] = df_handling['Price'].str.replace('€', '')
df_handling['Address'] = df_handling['Address'].str.replace('Porto ', '')
df_handling['Address'] = df_handling['Address'].str.replace('Lisbon ', '')
df_handling['Address'] = df_handling['Address'].str.replace('Braga ', '')
df_handling['Address'] = df_handling['Address'].str.replace('-', '')
df_handling['City'] = df_handling['Address'].str.split(',', expand=True)[0]
df_handling['Division'] = df_handling['Address'].str.split(',', expand=True)[1]
df_handling['City'] = df_handling['City'].str.strip()  # removing leading and tailing white spaces
df_handling['Size'] = df_handling['Size'].str.replace('m2', '')


# delete the Address column 

df_handling.drop('Address', inplace=True, axis=1)
```

```python
df_handling.head()
```

```python
# convert to int/float

df_handling['Price'] = pd.to_numeric(df_handling['Price'], errors='coerce')
df_handling['Size'] = pd.to_numeric(df_handling['Size'], errors='coerce')
df_handling['Bath'] = pd.to_numeric(df_handling['Bath'], errors='coerce')
df_handling['Bed'] = pd.to_numeric(df_handling['Bed'], errors='coerce')



print(df_handling.dtypes)
```

```python
df_handling.to_csv(r'df_after_handling.csv', index = False, header=True)
```

```python
def count_duplicatives(df, col_name=None):
    if(col_name ==None):
        return df.duplicated().sum()
    
    return df.duplicated([col_name]).sum()
```
